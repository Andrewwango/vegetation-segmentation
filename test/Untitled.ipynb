```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
with open("index_dict.pkl", "rb") as f:
    index_dict = pickle.load(f)
```

```python
def save_time_seq(seq, title, name, interval=333, ylabel=""): #univariate sequence
    x = np.arange(len(seq))
    fig, ax = plt.subplots(nrows=1, ncols=1)
    ax.set(xlim=(x.min(), x.max()), ylim=(seq.min(), seq.max()), title=title, xlabel="Frame", ylabel=ylabel, yticklabels=[])
    line = ax.plot(x[0], seq[0], color='k', lw=2)[0]
    anim = FuncAnimation(fig, lambda i: line.set_data(x[:i], seq[:i]), interval=interval, frames=len(seq)-1)
    anim.save(f'{name}')
```

```python
plt.plot(index_dict["grass"])
```

```python
save_time_seq(index_dict["grass"], "Vegetation index of grass", "VI_grass.gif", interval=333, ylabel="VI_grass")
```

```python
save_time_seq(index_dict["vegetation"], "Vegetation index of vegetation", "VI_veg.gif", interval=333, ylabel="VI_vegetation")
```
